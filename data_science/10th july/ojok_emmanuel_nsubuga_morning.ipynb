```python
# replace using mean, median, mode
# calculate mean
m = df4['Calories'].mean()
# print mean
print(m)
df4['Calories'].fillna(m, inplace=True)
```

```python
# Set the option to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df4
```

```python
# empty cells, wrong format, wrong data, duplicate values (Core of data preparation and preprocessing)) 
# calculate median

med = df4['Calories'].median()
print(med)  

df4['Calories'].fillna(med, inplace=True)   
```

```python
# Exercise 1
# calculate mode
mod = df4['Calories'].mode()
print(mod)
```

```python
# wrong format that is hard to analyze (NaN)
# eg 20230112
# NaN row 5 contained NaN values and we filled with 
df4['Timestamp'].fillna('05/07/2023 09:45:13', inplace=True)
df4
```

```python
# wrong data
# Row 23,37 contained typo in data 23, 2300 typo, we changed from 2300 to 23
for x in df4.index:
    if df4.loc[x, "gender"] == "Marabou stock":
        df4.drop(x, inplace = True)

df4
```
