```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Control display of seaborn data - optional
pd.options.display.max_columns = 50
sns.set(style="darkgrid",rc={"axes.facecolor": "CAF1DE"})
```

```python
# load the UCE results dataset
df = pd.read_csv('UceResultsBySchool2011-2016.csv')
df
```

```python
# confirm data shape
df.shape
```

```python
# columns
df.columns
```

```python
# preparation and preprocessing or data cleaning
df.columns
```

```python
# remove trailing spaces
df.columns = df.columns.str.strip()
```

```python
# after removing trailing spaces
df.columns
```

```python
# replace white spaces in column names with underscores
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
# check for information about the dataset
df.info()
```

```python
# check for duplicates
for isduplicated in df.duplicated():
    if isduplicated:
        print('There are duplicates')
        df.drop_duplicates(inplace=True)
    else:
        print("No duplicates")
```

```python
# check for number of  null and NaN values in each rown of the dataset
# df.isnull().sum()
null_rows = df[df.isnull().any(axis=1)]
before_computation = len(null_rows.axes[0])

print('Number of rows that are null: ', before_computation)
```

```python
# remove rows without school names
df.dropna(thresh=None, subset=['SCHOOL'], inplace=True)
```

```python
# print after removing null rows
null_rows = df[df.isnull().any(axis=1)]
after_computation = len(null_rows.axes[0])

print('Number of rows that are null: ', after_computation)
```
