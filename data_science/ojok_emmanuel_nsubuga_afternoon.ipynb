```python
# Define the dictionary
dates_dict = {
    'Monday': '2023-07-03',
    'Tuesday': '2023-07-04',
    'Wednesday': '2023-07-05',
    'Thursday': '2023-07-06',
    'Friday': '2023-07-07'
}

# Create the Pandas Series
series = pd.Series(dates_dict)

# Print the series
print(series)
```

```python
# Series using data from monday and tuesday
# Select data for Monday and Tuesday from the dictionary
monday_data = dates_dict['Monday']
tuesday_data = dates_dict['Tuesday']

# Create the Pandas Series
series = pd.Series([monday_data, tuesday_data], index=['Monday', 'Tuesday'])

# Print the series
print(series)
```

```python
# Read from csv file
# read from work.csv
work_pdf = pd.read_csv("work.csv")
```

```python
# head() method
# top 5 rows
work_pdf.head()
```

```python
# tail() method
# bottom 5 rows
work_pdf.tail()
```

```python
# info() method
# information about the data
work_pdf.info()
```

```python
# describe() method
# description of the data
work_pdf.describe()
```

```python
# fillna() method
# fill the missing values with 0
work_pdf_filled = work_pdf.fillna(0)
work_pdf_filled
```

```python
# dropna() method
# drop the missing values
work_pdf_filled.dropna()
```

```python
# drop duplicates
work_pdf_duplicate_dropped = work_pdf_filled.drop_duplicates()

# drop name and city columns
work_pdf_columns_dropped = work_pdf_duplicate_dropped.drop(['Name', 'city'], axis=1)

# print data with the columns dropped
work_pdf_columns_dropped

# Convert 'age' to numeric and drop rows with non-numeric values
work_pdf_columns_dropped['age'] = pd.to_numeric(work_pdf_columns_dropped['age'], errors='coerce')

# drop rows that do not have numeric values
work_pdf_columns_dropped_age_dropped = work_pdf_columns_dropped.dropna(subset=['age'])

# this data frame is one with rows in age without numerical values removed
work_pdf_columns_dropped_age_dropped

# Convert 'age' to numeric and drop rows with non-numeric values
work_pdf_columns_dropped_age_dropped['py-score'] = pd.to_numeric(work_pdf_columns_dropped_age_dropped['py-score'], errors='coerce')

# drop rows that do not have numeric values
work_pdf_columns_dropped_pyscore_dropped = work_pdf_columns_dropped_age_dropped.dropna(subset=['py-score'])

# this data frame is one with rows in age and py-score without numerical values removed
work_pdf_columns_dropped_pyscore_dropped

# we can now perform correlation
work_pdf_columns_dropped_pyscore_dropped.corr()
```
